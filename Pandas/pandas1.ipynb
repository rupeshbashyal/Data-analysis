```python
import pandas as pd
```

```python

numbers = range(1,100,5)
a=pd.Series(numbers)
print(a)


```

```python
# creating a pandas series with string values

str=pd.Series(['hello','how ','are','you'])
print(str)
```

```python
# creating a pandas series with different data types
hom= pd.Series([12,'butwal',3.5,'nepal',True], index=['age','city','height','country','is_student'])
print(hom)
```

```python
# set an index to the series
marks = [85, 90, 78, 92, 88]
sub = ['Math', 'Science', 'English', 'History', 'Art']
student_marks = pd.Series(marks, index=sub)
print(student_marks)
```

```python
# creating a pandas series from a dictionary

data ={'math':85, 'science':90, 'english':78, 'history':92, 'art':88}
student_marks = pd.Series(data)
print(student_marks)
```

```python
# null values in pandas series
subjects = ['math', 'science', 'social', 'history', 'Art']
ser= pd.Series(data, index=subjects)
print(ser)
```

```python
# check the numm value
print(ser.isnull())

# check the not numm value
print(ser.notnull())
```

```python
# filtering the series
print(ser[ser>60])
```

```python
# assigning values to null values

ser['social'] = 75
ser['Art'] = 88
print(ser)
print(ser.rank(ascending=False))
```

```python
import numpy as np
short_val = pd.Series([1,2,np.nan,4,5,np.nan,7,8,np.nan,10])
print(short_val)
print()
print(short_val.sort_values())
print()
print(short_val.rank())

```

#DataFrame


```python
# creating a pandas DataFrame from a dictionary

data = {
    "subjects":["math","science","arts"],
    "marks":(85,90,88),
    "CGPA":[3.5,3.8,3.6]
}
print(pd.DataFrame(data, index=['stu1','stu2','stu3']))
```

```python
# creating dataframe from series

subject = pd.Series(['math','science','arts'])
marks = pd.Series([85,90,88])
cgpa = pd.Series([3.5,3.8,3.6])

dat=pd.DataFrame([subject,marks,cgpa], index=['Subject','Marks','CGPA'])
print(dat)
print()

# converting into vertical format

ver=pd.DataFrame([subject,marks,cgpa], index=['Subject','Marks','CGPA']).T
print(ver)
```
