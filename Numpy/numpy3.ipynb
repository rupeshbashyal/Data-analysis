```python

# column supermarket: id, 2021, 2022, 2023, 2024

sales_data = np.array([
    [1, 12000, 13500, 15000, 17000], # kathmandu
    [2,  9000, 11000, 13000, 16000], # butwal
    [3, 15000, 14500, 15500, 18000], # pokhara
    [4,  8000,  9500, 12000, 14000], # biratnagar
    [5, 20000, 22000, 25000, 30000], # palpa
    [6, 17500, 18000, 19000, 21000]  # dharan
])
# print all the sales except ktm, only year 2022 and 2023
print(sales_data[1:,2:4])

print()


print ("--super market sales analysis--")
print (sales_data.shape)

# print first 3 
print (sales_data[0:3])
print()

# remove last column
print (sales_data[:,:4])
print()
#print column for 2023 and 2024
print (sales_data[:,2:4])
```

```python
# total sales per year

print (np.sum(sales_data[:,1:5], axis=0))  # column wise sum


#minimum sales in each city in 4 years
print (np.min(sales_data[:,1:5], axis=1))  # row wise min

# maximum sales in each city in 4 years
print (np.max(sales_data[:,1:5], axis=1))  # row wise max
```

```python
# averaage sales in each city in 4 years
print (np.mean(sales_data[:,1:], axis=1))  # row wise mean
```
