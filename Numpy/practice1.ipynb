```python
import numpy as np
```

```python

## creating a numpy array

data =np.array([1,2,3,4,5])
print(data)

array2d = np.array([[1,2,3],[4,5,6]])
print(array2d)
```

```python

array1 = np.array((1,2,3,4,5))
array2 = np.array((10,20,30,40,50))

print ("multiplication of two arrays:",array1*array2)
```

```python

num = np.arange(0,10)
print(num)

##creating array with one

one = np.ones((4,5))
print(one)

## crating array with random

random = np.random.random((4,5))
print(random)
```

```python
##vector , matrix and tensor

v1 = np.array([101,102,103])
print("vector:",v1)

m1 =np.array([[1,2,3],[11,12,13],[21,22,23]])
print("matrix:",m1)

t1 = np.array([[[1,2,3],[1,2,3]],[[4,5,6],[4,5,6]]])
print("tensor:",t1)
print("shape of tensor:",t1.shape)
print("dimension of tensor:",t1.ndim)
```

```python

dta =np.array([[1,2,3],[11,12,13],[21,22,23]])
print("shape:",dta.shape)
print("dimension:",dta.ndim)
print("size",dta.size)
print("data type",dta.dtype)

```

```python
array = np.arange(1,16)
print("\n",array)

reshapearray = array.reshape(3,5)
print("\n",reshapearray)


flattenarray = reshapearray.flatten()
print("\n",flattenarray)

raveled = reshapearray.ravel() ## shows the orginal array instead of copy
print("\n",raveled)

transposearray = reshapearray.T
print("\n",transposearray)
```

```python
a1 = np.array([[1,2,3,4],[5,6,7,8]])

m = a1.mean()
print("mean:",m)

sta = a1.std()
print("standard deviation:",sta)

```
