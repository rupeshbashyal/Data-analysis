```python
# 1Given:
# a = np.array([2, 4, 6])
# b = np.array([1, 3, 5])

# a.Add, subtract, multiply, and divide the arrays
# b.Find a²

a = np.array([2, 4, 6])
b = np.array([1, 3, 5])

print("Adding the arrays:", a + b)
print("Subtracting the arrays:", a - b)
print("Multiplying the arrays:", a * b)
print("Dividing the arrays:", a / b)
print("Squaring array a:", a**2)
```

```python
# 2.Find the sum, mean, min, max, and standard deviation of:

# arr = np.array([10, 20, 30, 40, 50])


arr = np.array([10, 20, 30, 40, 50])

print("Sum:", arr.sum())
print("Mean:", arr.mean())
print("Min:", arr.min())
print("Max:", arr.max())
print("Standard Deviation:", arr.std())
```

```python
# 3.Find the row-wise sum and column-wise sum of:

# arr = np.array([[1, 2, 3],
#                 [4, 5, 6]])

arr = np.array([[1, 2, 3],
                [4, 5, 6]])

print("sum according to column",np.sum(arr,axis=0))
print("sum according to row",np.sum(arr,axis=1))
```

1.From the array:
arr = np.array([12, 7, 9, 15, 20, 3])

a.Select values greater than 10
b.Replace values less than 10 with 0

2.Given:

scores = np.array([45, 67, 89, 34, 76, 90])


a.Find students who passed (≥ 50)

b.Count how many students failed

```python
# 1.From the array:
# arr = np.array([12, 7, 9, 15, 20, 3])

# a.Select values greater than 10
# b.Replace values less than 10 with 0


arr = np.array([12, 7, 9, 15, 20, 3])

print("Values greater than 10:",arr[arr>10])
print()
arr[arr<10] = 0
print("replacing value with 0 which is less than 10",arr)
    
```

```python
# 2.Given:

# scores = np.array([45, 67, 89, 34, 76, 90])


# a.Find students who passed (≥ 50)

# b.Count how many students failed

scores = np.array([45, 67, 89, 34, 76, 90])

print("Student who have passed :",scores[scores >= 50])

print("Number of students who failed:", np.count_nonzero(scores <= 50))
```

1.Sales Analysis
sales = np.array([1000, 1200, 900, 1500, 1800, 1100])

a.Find total sales
b.Find average sales
c.Find the highest and lowest sales

2.2D Sales Data
sales = np.array([[200, 300, 250],
                  [400, 500, 450],
                  [600, 700, 650]])

a.Find total sales per day
b.Find total sales per product
c.Find the overall total sales

3.Replace all values below the average with the average value in the array:
arr = np.array([10, 20, 30, 40, 50])

```python
# 1.Sales Analysis
# sales = np.array([1000, 1200, 900, 1500, 1800, 1100])

# a.Find total sales
# b.Find average sales
# c.Find the highest and lowest sales
sales = np.array([1000, 1200, 900, 1500, 1800, 1100])

print("Total sales:", np.sum(sales))
print("Average sales:", np.mean(sales))
print("Lowest sales:", np.min(sales))
print("Highest sales:", np.max(sales))
```

```python
# 2.2D Sales Data
# sales = np.array([[200, 300, 250],
#                   [400, 500, 450],
#                   [600, 700, 650]])

# a.Find total sales per day
# b.Find total sales per product
# c.Find the overall total sales


sales = np.array([[200, 300, 250],
                  [400, 500, 450],
                  [600, 700, 650]])

day_sales=np.sum(sales, axis=1)
print("Total sales for each day:",day_sales)
product_sales = np.sum(sales, axis=0)
print("Total sales for each product:",product_sales)

print("Total sales for all products:",np.sum(sales))

```

```python
# 3.Replace all values below the average with the average value in the array:
# arr = np.array([10, 20, 30, 40, 50])


arr = np.array([10, 20, 30, 40, 50])
avg=np.mean(arr)
arr[arr<avg] = avg
print(arr)

```

1.Create an array of even numbers from 2 to 100
2.Reverse the array

```python

# 1.Create an array of even numbers from 2 to 100
array= np.array([])
for i in range(2,102):
    if i % 2 ==0:
        array= np.append(array,i)
print("Array from 1 to 100 contains only even numbers:",array)
print()

# 2.Reverse the array
reverse = array[::-1]
print("Reverse the array:",reverse)

# print(np.flip(array))
```
