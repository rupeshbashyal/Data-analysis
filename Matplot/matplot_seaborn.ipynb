```python
import matplotlib.pyplot as ptl
import seaborn as sns
```

```python
# example dataset in seaborn
Example_dataset = sns.load_dataset('tips')
Example_dataset.head()
```

```python
Example_dataset.info()
```

```python
import numpy as np
```

```python
x= np.linspace(1,20,100)
y = np.exp(x)
# x and y must have same value to plot, if it hass different value matplot  doesnot which x matches which y
ptl.plot(x,y)
ptl.show()

```

```python
# ploting each point with red colour with dot'.'
ptl.plot(x,y, 'r.')
ptl.show()

```

```python
# plotting multiple lines

x= np.linspace(1,20,50)
a1=x
a2= np.square(x)
a3= np.sqrt(x)

ptl.plot(x,a1,'r',x,a2,'g',x,a3,'b') #each line has its own value of x
ptl.title("Line Plot")
ptl.show()
```

```python
print(Example_dataset.head())
x=Example_dataset["total_bill"]
y=Example_dataset["tip"]

ptl.scatter(x,y)
ptl.xlabel("Total Bill") #adding info to the plot
ptl.ylabel("Tips")
ptl.show()
```

```python
# adding different colour, opacity and shape to the data points
x=Example_dataset["total_bill"]
y=Example_dataset["tip"]

ptl.scatter(x,y, s=np.array(y)**2,c="green",alpha=0.8)
ptl.xlabel("Total Bill")
ptl.ylabel("Tips")
ptl.show()
```

```python
import pandas as pd
data= pd.read_csv("Dataset.csv")
data.head()


x= data["age"]
y= data["salary"]

ptl.scatter(x,y,s=np.array(x)**1.2,c="blue",alpha=0.8)
ptl.title("salary and age")
ptl.xlabel("Age")
ptl.ylabel("Salary")
ptl.show()
```

```python
# Bar Plot
from matplotlib.pyplot import bar


set(data["gender"])

data.gender.value_counts().plot(kind="bar",rot=0)
ptl.show()
```

```python
# set(data["department"])

print(data.department.value_counts())

data.department.value_counts().plot(kind='bar',rot=0, color ='red')
ptl.title("Departments in our company")
ptl.ylabel("Counts")
ptl.show()
```

```python
# pie chart
print(data.department.value_counts())

ptl.pie(data.department.value_counts(),autopct='%.1f%%', radius=1,labels=["HR","Finance","marketing","IT","sales"])

ptl.show()
```

```python
ptl.pie(data.department.value_counts(),autopct='%.1f%%',
 radius=1,labels=["HR","Finance","marketing","IT","sales"],explode=[0,0,0.2,0,0])

ptl.show()
```

```python
# Box plot
ptl.boxplot(data["salary"])

ptl.show()
print(data.salary.describe())
```

```python
sns.stripplot(y=data['performance_score'], x=data['gender'], jitter= False)
ptl.show()
```

```python
# swarm plot

# swarm plot is used for better understing of the data.
sns.swarmplot(y=data['experience_years'],x=data['gender']) 
ptl.show()
```

```python
sns.swarmplot(y=data['experience_years'],x=data['gender'],hue=data['department']) 
ptl.show()
```

```python
ptl.figure(figsize=(1,1))
sns.pairplot(data)
ptl.show()
```

```python
# distribution plot

sns.histplot(data['salary'], kde=True)
ptl.show()
```
